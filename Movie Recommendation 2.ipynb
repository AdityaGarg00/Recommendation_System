```python
import pandas as pd
import webbrowser as wb
```

```python
import numpy as np
import matplotlib.pyplot as mtp
import difflib #if movie name is incrorect it will correct it
from sklearn.feature_extraction.text import TfidfVectorizer # Convert textual data into numbers
from sklearn.metrics.pairwise import cosine_similarity # similarity score between movies
```

```python
data = pd.read_excel("Netflix.xlsx")
```

```python
data
```

```python
feature = ['Genre', 'Tags', 'Director']
```

```python
for features in feature:
    data[feature] = data[feature].fillna(" ")
```

```python
combine = data['Genre']+ ''+data['Tags']+'' +data['Director']
combine
```

```python
vector = TfidfVectorizer()
feature_vector = vector.fit_transform(combine)
print(feature_vector)
```

```python
similarity = cosine_similarity(feature_vector)
```

```python
similarity
```

```python
list_of_all_movies_title = data['Title'].tolist()
print(list_of_all_movies_title)
```

```python
index_of_movie = int(data[data.Title == movie_name_input]['index'].values[0])
```

```python
index_of_movie
```

```python
similarity_score = list(enumerate(similarity[index_of_movie]))
```

```python
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
```

```python
sorted_similar_movies
```

```python
print('Movies Suggested for you : \n' )
i = 1
index_list_of_movie=[]
for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = data[data.index == index]['Title'].values[0]
    index_value_result = data[data.index == index]['index'].values[0]
    if(i< 10):
        index_list_of_movie.append(index_value_result)
        print(i, ".", title_from_index) 
        i +=1
```

```python
ll = []
for i in range(0, 9):
    num1 = index_list_of_movie[i]
    num2 = int(num1)
    ll.append(num2)
```
